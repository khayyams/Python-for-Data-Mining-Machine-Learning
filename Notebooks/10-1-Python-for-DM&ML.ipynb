```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data

```python
df = pd.read_csv("./data/Advertising.csv")
df.head()
```

```python
df.drop("Unnamed: 0", axis = 1, inplace=True)
df.head()
```

## Train-Test Split Framework:
1. Pre-processing step

2. Splitting into train and test sets:

3. Scaling data

4. Importing model algorithms:

5. Creating an instance of the model:
6. Training the model
7. Predicting the $\hat{y}$
8. Importing error metrics
9. Comparing with y_test
10. Adjust hyperparameters as necessary and repeat steps 5 to 9

```python
X = df.drop('sales',axis=1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=32)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha=100)
```

```python
model.fit(X_train,y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_predicted)
```

**Adjusting hyperparameter**

```python
model = Ridge(alpha=1)
```

```python
model.fit(X_train,y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
mean_squared_error(y_test,y_predicted)
```

## Train-Validation-Test Split Framework:
1. Pre-processing step
2. Splitting into train, validation, and test sets:
3. Scaling data
4. Importing model algorithms
5. Creating an instance of the model
6. Training the model
7. Predicting the $\hat{y}$
8. Importing error metrics
9. Comparing with y_test 
10. Adjust hyperparameters as necessary and repeat steps 5 to 9
11. report final metrics on Test set (not allowed to go back and adjust after this!)

```python
X = df.drop('sales',axis=1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split

# 70% of data is training data, set aside other 30%
X_train, X_OTHER, y_train, y_OTHER = train_test_split(X, y, test_size=0.3, random_state=32)

# Remaining 30% is split into evaluation and test sets in the same size
X_eval, X_test, y_eval, y_test = train_test_split(X_OTHER, y_OTHER, test_size=0.5, random_state=32)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_eval = scaler.transform(X_eval)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha=100)
```

```python
model.fit(X_train,y_train)
```

```python
y_eval_predicted = model.predict(X_eval)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_eval,y_eval_predicted)
```

**Adjusting hyperparameters**

```python
model = Ridge(alpha=1)
```

```python
model.fit(X_train,y_train)
```

```python
y_eval_predicted = model.predict(X_eval)
```

```python
mean_squared_error(y_eval,y_eval_predicted)
```

```python
y_final_test_predicted = model.predict(X_test)
```

```python
mean_squared_error(y_test,y_final_test_predicted)
```

## Cross Validation with cross_validate

* specifying multiple metrics for evaluation.

* returning a dict containing fit-times, score-times (and optionally training scores as well as fitted estimators) in addition to the test-score.

```python
X = df.drop('sales',axis=1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=32)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Ridge(alpha=100)
```

```python
from sklearn.model_selection import cross_validate
```

```python
scores = cross_validate(model, X_train, y_train,
                        scoring=['neg_mean_absolute_error','neg_mean_squared_error','max_error'],
                        cv=5)
```

```python
scores
```

```python
pd.DataFrame(scores)
```

```python
pd.DataFrame(scores).mean()
```

**Adjusting hyperparameters based on metrics**

```python
model = Ridge(alpha=1)
```

```python
scores = cross_validate(model, X_train, y_train,
                        scoring=['neg_mean_absolute_error','neg_mean_squared_error','max_error'],
                        cv=5)
```

```python
pd.DataFrame(scores).mean()
```

```python
# Need to fit the model first!
model.fit(X_train,y_train)
```

```python
y_final_test_predicted = model.predict(X_test)
```

```python
mean_squared_error(y_test,y_final_test_predicted)
```

# Grid Search 

* search through a variety of combinations of hyperparameters with a grid search

`GridSearchCV` search for:

* an estimator (e.g. a regressor or classifier);
* a parameter space;
* a method for searching or sampling candidates;
* a cross-validation scheme 
* a score function.

```python
df = pd.read_csv("./data/Advertising.csv")
df.drop("Unnamed: 0", axis = 1, inplace=True)
df.head()
```

```python
X = df.drop('sales',axis=1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=32)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
help(ElasticNet)
```

```python
base_elastic_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.01, 0.1, 1, 5, 10, 50],
              'l1_ratio':[0.01, 0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_elastic_model,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error',
                          cv=5,
                          verbose=2)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
grid_model.cv_results_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_predicted = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_predicted)
```
