```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Scatter Plots

* can show the relation of different features,
* can show the interconnection of features.

```python
df = pd.read_csv("./data/sales.csv")
df.head()
```

```python
df.info()
```

## `scatterplot()`

```python
sns.scatterplot(x='salary',y='sales',data=df)
```

* matplotlib is connected to seaborn underneath.
* seaborn is directly making a Figure call with matplotlib.

=> We can import matplotlib.pyplot and make calls to directly effect the seaborn figure.

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df)
```

### hue
* coloring categorical features

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df, hue='division')
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience')
```

### palette

https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience',palette='magma')
```

### points size and transparency

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=50)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=100,alpha=0.4)
```

### size
* resize points based on another column

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,size='work experience')
```

### style

* Automatically choosing styles based on another categorical feature
* Optionally using the **markers=** parameter to pass a list of marker choices, for example: ['*','+','o']

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education')
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',\
                hue='level of education',s=50)
```

```python
plt.figure(figsize=(12,8), dpi=300)
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',\
                hue='level of education',s=75)

plt.savefig('scatter.png')
```

## Distributions of a feature

* rugplot
* distplot (histogram)
* kdeplot.

### `rugplot()`

* one mark per data point
* not so usable (especially for large data sets)

```python
sns.rugplot(x='salary',data=df)
```

```python
plt.figure(figsize=(12,8), dpi=200)
sns.rugplot(x='salary',data=df, hue="level of education", height = 0.3)
```

### `displot()` and `histplot()`

**Instead of plotting data points on an axis, count the number of points per some segment of the axis**

=>histogram

`displot()` is a histogram with the option of adding a "KDE" plot (Kernel Density Estimation) on top of the histogram.

```python
sns.displot(data=df,x='salary')
```

```python
sns.displot(data=df,x='salary',kde=True)
```

```python
sns.histplot(data=df,x='salary')
```

```python
sns.histplot(data=df,x='salary',bins=15)
```

```python
sns.histplot(data=df,x='salary',bins=100)
```

**Styles:**
* darkgrid
* whitegrid
* dark
* white
* ticks

```python
sns.set(style='darkgrid')
sns.histplot(data=df,x='salary',bins=100)
```

```python
sns.set(style='white')
sns.histplot(data=df,x='salary',bins=100)
```

**Adding in keywords from matplotlib**

* the main styling parameters

```python
sns.displot(data=df,x='salary',bins=20,kde=False, \
            color='red',edgecolor='black',lw=4,ls='--')
```

### `barplot()`

* in `countplot()`, y axis represents a count. 
* other functions?

```python
plt.figure(figsize=(10,6),dpi=200)
sns.barplot(x='level of education',y='salary',data=df,ci='sd')
```

```python
plt.figure(figsize=(10,6),dpi=200)
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci=None)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci=None,hue='division')
```

```python
plt.figure(figsize=(12,6),dpi=100)

sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci=None,hue='division')

plt.legend(bbox_to_anchor=(1.05, 1))
```

## Grids

### `pairgrid()`

```python
g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot, lw=2)
g = g.map_lower(sns.kdeplot, colors="red")
```

```python
g = sns.PairGrid(df, hue="gender", palette="viridis",hue_kws={"marker": ["o", "+"]})
g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.distplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend();
```

## The Data

https://www.ined.fr/en/everything_about_population/data/all-countries/?lst_continent=900&lst_pays=926

```python
# 2020 Projections
df = pd.read_csv('./data/countries.csv')
df.head()
```

### `heatmap()`

```python
df = df.set_index('Countries')
df.head()
```

```python
sns.heatmap(df)
```

```python
rates = df.drop('Total population (thousands)',axis=1)
rates = rates.drop('Population aged 65 and more (thousands)',axis=1)
rates.head()
```

```python
sns.heatmap(rates)
```

```python
rates = rates.drop('Life expectancy',axis=1)
rates.head()
```

```python
sns.heatmap(rates)
```

```python
sns.heatmap(rates,linewidth=0.5)
```

```python
sns.heatmap(rates,linewidth=0.5,annot=True,cmap='viridis')
```

```python
# Set colorbar based on value from dataset
sns.heatmap(rates,linewidth=0.5,annot=True,cmap='viridis',center=20)
```

```python
# Set colorbar based on value from dataset
sns.heatmap(rates,linewidth=0.5,annot=True,cmap='viridis',center=40)
```

### `clustermap()`

Plotting as a hierarchically-clustered heatmap.

```python
sns.clustermap(rates)
```

```python
sns.clustermap(rates,col_cluster=False)
```

```python
sns.clustermap(rates,col_cluster=False,figsize=(12,8),cbar_pos=(-0.1, .2, .03, .5))
```
